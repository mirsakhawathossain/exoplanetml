```python
#import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#define a function to calculate VIF
def vif(df):
  #initialize an empty dataframe to store VIF values
  vif_df = pd.DataFrame()
  #get the names of the features
  vif_df["feature"] = df.columns
  #loop through each feature and regress it against the others
  vif_df["VIF"] = [1 / (1 - LinearRegression(n_jobs=-1).fit(df.drop(feature, axis=1), df[feature]).score(df.drop(feature, axis=1), df[feature])) for feature in df.columns]
  #return the VIF dataframe
  return vif_df
```

```python
exoplanet_data = pd.read_csv('https://gitlab.com/mirsakhawathossain/exodata/-/raw/main/dataset/exodata.csv').sort_index(axis=0)
```

```python
drop_columns = exoplanet_data.columns[exoplanet_data.mean()==0].to_list()+exoplanet_data.columns[exoplanet_data.mean()==1].to_list()
```

```python
#calculate VIF for each feature
vif_df = vif(df)

#print the VIF dataframe
print(vif_df)
```
