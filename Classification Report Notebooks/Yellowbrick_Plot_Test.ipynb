# Yellowbrick Plot Test

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
# import seaborn as sns
# sns.set()
mpl.rcParams["figure.figsize"] = (10, 10)
mpl.rcParams["axes.grid"] = True
mpl.rcParams["font.size"] = 16
```

```python
from yellowbrick.datasets import load_hobbies
from yellowbrick.text.correlation import WordCorrelationPlot

# Load the text corpus
corpus = load_hobbies()

# Create the list of words to plot
words = ["Tatsumi Kimishima", "Nintendo", "game", "play", "man", "woman"]

# Instantiate the visualizer and draw the plot
# viz = WordCorrelationPlot(words,size=(1080, 720))
viz = WordCorrelationPlot(words,size=(1080, 720))
viz.fit(corpus.data)
# for label in viz.ax.texts:
#    label.set_size(36)
viz.show(outpath="test.png")
```

```python
# Get access to the axes object and modify labels
viz.ax.set_xlabel("measured concrete strength")
viz.ax.set_ylabel("predicted concrete strength")
plt.savefig("peplot.pdf")
```
